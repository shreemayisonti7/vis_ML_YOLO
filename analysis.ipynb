```python
from ultralytics import YOLO
import os
```

```python
model = YOLO("yolov5n.pt")

# Folder name of images to run inference
image_folder = 'right_park2_sensor2_iframes'

#0: 'person', 1: 'bicycle', 2: 'car', 3: 'motorcycle', 4: 'airplane', 5: 'bus', 6: 'train', 7: 'truck'
set_R = set([0, 1, 3])
set_D = set([2, 5, 7])

filenames = sorted(os.listdir(image_folder))
len(filenames)
```

```python
heavy_vehicle_count = 0
risk_count = 0
light_vehicle_count = 0
for file_name in filenames:
    results = model(f'{image_folder}/{file_name}')
    result = results[0]
    curr_set = set(result.boxes.cls.tolist())
    if curr_set.intersection(set_D):
        heavy_vehicle_count += 1
        if curr_set.intersection(set_R):
            risk_count += 1
    elif curr_set.intersection(set_R):
        light_vehicle_count += 1
```

```python
print(heavy_vehicle_count, risk_count, light_vehicle_count)
```
